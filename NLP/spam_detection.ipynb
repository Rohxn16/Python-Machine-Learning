```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('datasets/spam.csv')
df.head()
```

```python
df.Category.value_counts() #utility to get the number of each kind of selected column
```

```python
df['spam'] = df['Category'].apply(lambda x: 1 if x == 'spam' else 0) 
# translating the above line to english, it stands for create a new col spam in df for every value in category and apply a transformation that maps each value
# to 0 or 1 based on whether they are spam or not

#the above lambda expression can be replaced by a simple function like this, lambda is used for ease
def getSpamNumber(spam):
    if spam == 'spam':
        return 1
    return 0
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.Message, df.spam, test_size=0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

v = CountVectorizer()
X_train_cv = v.fit_transform(X_train.values)
X_train_cv.toarray()[:5]
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train_cv,y_train)
```

```python
#for testing we need to convert the test set into a vectorized set first
X_test_cv = v.transform(X_test)
```

## Here we are using classification report instead of accuracy because the dataset is imbalanced.

```python

from sklearn.metrics import classification_report

y_pred = model.predict(X_test_cv)

print(classification_report(y_test,y_pred))

```

## Alternate (easier) method of doing the above ML model using sklearn pipeline

```python
from sklearn.pipeline import Pipeline

clf = Pipeline([
    ('vectorizer',CountVectorizer()),
    ('nb',MultinomialNB())
])
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
print(classification_report(y_test,y_pred))
```
