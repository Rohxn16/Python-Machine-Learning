# This is gonna be a ground up implementation of the Sigmoid function for logistic regression
### In a linear funciton we have seen $f_w,_b(x^{(i)}) = w . x^{(i)} + b$ which gives us a number as prediciton, however in classification models we need our predictions to be in either true(1) or false(0) which can be achieved by using a sigmoid function that can map all the values given to it between 0 and 1.

## Formula for Sigmoid function:
### g(z) = $\frac{1}{1+e^{-z}}$

- in the case of a single example,  ùëß
  is scalar.
- in the case of multiple examples,  ùëß
  may be a vector consisting of  ùëö
  values, one for each example.

```python
#Implemeting the sigmoid function
input_array = np.array([1,2,3])
exp_array = np.exp(input_array)
print(input_array)
print(exp_array)
```

```python
def sigmoid(z):
#     z is an np array of size m
    g = 1/(1+np.exp(-z))
    return g
```

```python
# generating sample sigmoid values for a sigmoid array
z_temp = np.arange(-10,11)
y_temp = sigmoid(z_temp)
m = z_temp.shape[0]
np.set_printoptions(precision=3)
print(np.c_[z_temp,y_temp])
    
```
